### Convert dataset to allow kaggle happiness to ensue

```python
import pandas as pd
import os, sys

# get datafile name and directory
dir_curr = os.getcwd()
dir_data = os.path.join(dir_curr, '../data')

for i in os.listdir(dir_data):
    data_file = i
    
# load dataset into memory
df_0 = pd.ExcelFile(os.path.join(dir_data,data_file))
```

```python
# name all tabs/sheet in file
data_sheets = df_0.sheet_names

# read data in sheet-wise
df_0910 = df_0.parse(data_sheets[0])
df_1011 = df_0.parse(data_sheets[1])

# merge dataframes into single master
df_0911 = df_0910
df_0911 = df_0911.append(df_1011)
```

```python
conv_dict = {}
for i in df_0911:
    conv_dict[i] = 'str'
    
conv_dict
```

```python
df_test = df_0911.astype(conv_dict)
```

```python
# write master to csv
df_0911.to_csv(os.path.join(dir_data,'online_retail_20092011.csv'), encoding='utf-8')
```
